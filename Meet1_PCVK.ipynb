```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # import cv2_imshow from google.colab.patches

# Konversi gambar dari BGR ke RGB
image_2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_3 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Dapatkan dimensi gambar
tinggi, lebar = image.shape[:2]

# Fungsi untuk menggambar garis diagonal dari titik (x1, y1) ke titik (x2, y2)
def draw_diagonal_line(img, x1, y1, x2, y2, color):
    dx = x2 - x1
    dy = y2 - y1
    steps = max(abs(dx), abs(dy))

    x_increment = dx / float(steps)
    y_increment = dy / float(steps)

    x = x1
    y = y1

    for _ in range(int(steps) + 1):
        if 0 <= int(x) < img.shape[1] and 0 <= int(y) < img.shape[0]:
            img[int(y), int(x)] = color
        x += x_increment
        y += y_increment

# Menggambar garis diagonal dari kiri atas ke kanan bawah
draw_diagonal_line(image_3, 0, 0, lebar - 1, tinggi - 1, [255, 255, 255])

# Menggambar garis diagonal dari kiri bawah ke kanan atas
draw_diagonal_line(image_3, 0, tinggi - 1, lebar - 1, 0, [255, 255, 255])

# Menggabungkan dua gambar secara horizontal
final_frame = cv2.hconcat([image_2, image_3])

# Menampilkan hasil
cv2_imshow(final_frame)  # Gunakan cv2_imshow daripada cv2.imshow
cv2.waitKey(0)
cv2.destroyAllWindows()
```
